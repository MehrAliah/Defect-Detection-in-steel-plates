```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import cv2
import matplotlib.pyplot as plt


# Load the saved model
model = tf.keras.models.load_model(r"D:\Internship\models\model.h5")


# Set the parameters for image classification
batch_size = 16
img_size = 150
img_path = r"D:\Internship\archive\validation\inclusion\inclusion_242.jpg"
# Preprocess the image
test_datagen = image.ImageDataGenerator(rescale=1./255)

# Load and preprocess the image
img = image.load_img(img_path, target_size=(img_size, img_size), color_mode='grayscale')
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0  # Normalize the image pixel values

# Display the image
plt.imshow(img)
plt.axis('off')
plt.show()

# Make the prediction
class_mapping = {0: 'crazing', 1: 'inclusion', 2: 'patches', 3: 'pitted_surface', 4: 'rolled-in_scale', 5: 'scratches'}
prediction = model.predict(img_array)
predicted_class = np.argmax(prediction)
predicted_class_name = class_mapping[predicted_class]

# Print the predicted class
print("Predicted class:", predicted_class_name)
print("index          :", predicted_class)
```
